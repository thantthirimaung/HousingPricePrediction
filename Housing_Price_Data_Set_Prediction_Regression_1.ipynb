### 2. Data Preprocessing

```python
# null values တွေကို ဖယ်တဲ့ Code

#Note: By default, the dropna() method returns a new DataFrame, and will not change the original.
# If you want to change the original DataFrame, use the inplace = True argument:

train_data.dropna(inplace = True)
test_data.dropna(inplace= True)
```

```python
train_data.iloc[:,-1]
# train_data.iloc[:, 0:-1 ]
```

```python
# features X နဲ့  Labels Y ခွဲပေးတဲ့ Code,
# in short, feautures == x and labels == y
# features = input value that describes the characteristics of such labels in datasets
# label = output value

X_train = train_data.iloc[:,:-1].values
Y_train = train_data.iloc[:,-1].values

X_test = test_data.iloc[:,:-1].values
Y_test = test_data.iloc[:,-1].values

```

### 3. Modeling

```python
import sklearn
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# fitting the data
model.fit(X_train, Y_train)    #model.fit က for modelling
# Training Data
```

```python
#  Testing Data ရဲ့Features တွေပေးပြီး ခန့်မှန်းမယ်။
# predict the data
Y_predicted = model.predict(X_test)   #to test, give only features
```

```python
print(Y_predicted[0: 3])
```

```python
# မူလ ရှိတဲ့ Labels ကို ကြည့်ကြမယ်။
print(Y_test[0:3])
```

### 4. Evaluation

```python
# Mean Abs Error
total = 0
for i in range(len(Y_test)):
  abs_error = abs(Y_test[i]-Y_predicted[i])
  total += abs_error
mae = total/len(Y_test)
print("Mean Abs Error", mae)
```

```python
# Mean Square Error
total = 0
for i in range(len(Y_test)):
  abs_error = (Y_test[i]-Y_predicted[i])**2
  total += abs_error
mse = total/len(Y_test)
print("Mean square Error", mse)
```
